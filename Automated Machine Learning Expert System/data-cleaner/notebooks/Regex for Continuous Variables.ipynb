```python
def checkNumeric(num):
    if num == num:
        return True if re.search("[0-9]+.*[0-9]*", num) else False
    else:
        return False
```

```python
s="sales = +998327.812*112211"
if checkNumeric(s):
    res=""
    for i in s:
        res+=re.sub("[`~!@#$=+%^&*_/?()-]|[a-zA-Z]", "", i)
    res=float(res)
    print("Original =",s," Numeric =",res)
else:
    print("Not Numeric")
```
