## Experimental Results

```python
from pprint import pprint
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("../uploads/IrisCopy_-_Copy_Final_Encoded.csv")
df2 = pd.read_csv("../uploads/Updated_iris_-_Copy_Final_Encoded.csv")
```

```python
df1 = pd.read_csv("../uploads/Originaliris_Final_Encoded.csv")
df2 = pd.read_csv("../uploads/iris_Final_Encoded.csv")
```

```python
df3 = pd.read_csv("../uploads/iris.csv")
df4 = pd.read_csv("../uploads/Originaliris.csv")
```

```python
idx = {}

sepal_length = df3[df3['sepal length (cm)'].isna() == True].index.tolist()
sepal_width = df3[df3['sepal width (cm)'].isna() == True].index.tolist()

petal_length = df3[df3['petal length (cm)'].isna() == True].index.tolist()
petal_width = df3[df3['petal width (cm)'].isna() == True].index.tolist()

idx['sepal length (cm)'] = sepal_length
idx['sepal width (cm)'] = sepal_width
idx['petal length (cm)'] = petal_length
idx['petal width (cm)'] = petal_width

print(idx)
```

```python
df1
```

```python
d = pd.DataFrame(columns=['original', 'replaced'])

c=2
for i in idx.keys():
    for j in idx[i]:
        d.loc[d.shape[0]] = [df1.loc[j, str(c+1)], df2.loc[j, str(c+1)]]

print(d)
```

```python
print("root mean_squared_error         =", round(mean_squared_error(d['original'].values, d['replaced'].values)**0.5, 6))
print("r2_score                        =", round(r2_score(d['original'].values, d['replaced'].values), 6))
print("mean_absolute_error             =", round(mean_absolute_error(d['original'].values, d['replaced'].values), 6))
print("mean_absolute_percentage_error  =", round(mean_absolute_percentage_error(d['original'].values, d['replaced'].values), 6))
```
